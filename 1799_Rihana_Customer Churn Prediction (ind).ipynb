## Dataset Overview

The Telco Customer Churn dataset contains information on customers of a telecommunications company, including various attributes like:

- **CustomerID**: Unique identifier for each customer.
- **Gender**: Male or Female.
- **SeniorCitizen**: Indicates if the customer is a senior citizen (1) or not (0).
- **Partner**: Whether the customer has a partner (Yes/No).
- **Dependents**: Whether the customer has dependents (Yes/No).
- **Tenure**: Number of months the customer has been with the company.
- **PhoneService**: Whether the customer has phone service (Yes/No).
- **MultipleLines**: Whether the customer has multiple lines (Yes/No).
- **InternetService**: Type of internet service (DSL, Fiber optic, No).
- **OnlineSecurity**, **OnlineBackup**, **DeviceProtection**, etc.: Various optional services (Yes/No).
- **Contract**: Type of contract (Month-to-month, One year, Two year).
- **PaperlessBilling**: Whether the customer uses paperless billing (Yes/No).
- **PaymentMethod**: Payment method (Electronic check, Mailed check, Bank transfer, Credit card).
- **MonthlyCharges**: The monthly fee charged to the customer.
- **TotalCharges**: Total amount charged.
- **Churn**: Whether the customer hasble: `Churn` (Yes/No)

#### Data preprocessing :

```python
X=df.drop(columns=['Churn'])
y=df['Churn']
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()

cat_clmn=X.select_dtypes(include=['object']).columns

for col in cat_clmn:
    X[col]=le.fit_transform(X[col])

```

```python
X.head(1)
```

```python
def convert(x):
    if x == 'Yes':
        return 1
    elif x == 'No':
        return 0

y=y.apply(convert)
```

#### Standardize data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size =0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
X_train_scaled_df=pd.DataFrame(X_train_scaled,columns=X_train.columns)
X_test_scaled_df=pd.DataFrame(X_test_scaled,columns=X_test.columns)
X_train_scaled_df.head()
```

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr_model=LogisticRegression()
lr_model.fit(X_train_scaled,y_train)
```

```python
lr_model.score(X_train_scaled,y_train)
```

```python
lr_y_pred=lr_model.predict(X_test_scaled)
```

```python
from sklearn import metrics

print(f"Accuracy of prediction: {metrics.accuracy_score(y_test,lr_y_pred)}")
print("Confusion matrix: ")
print(metrics.confusion_matrix(y_test,lr_y_pred))
print("Classification report: ")
print(metrics.classification_report(y_test,lr_y_pred))
```

```python
sns.heatmap(metrics.confusion_matrix(y_test,lr_y_pred),annot=True,fmt='d')
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(),X,y,cv=5,scoring='accuracy').mean()
```
